```python
import random
import pandas as pd

age = [random.randint(20, 60) for i in range(15)]
gender = [random.choice(['M', 'F']) for index in range(15)]
occupation = [random.choice(['Engineer', 'Scientist', 'Teacher', 'Doctor', 'Business']) for index in range(15)]
city = [random.choice(['Stuttgart', 'Munich', 'Berlin', 'Cologne', 'Hamburg']) for index in range(15)]
garden_size = [random.randint(0,10) for i in range(15)]
current_planting = [j*random.randint(1, 10) for i, j in zip(range(15), garden_size)]
household_income = [random.randint(2000, 10000) for i in range(15)]
household_size = [random.randint(1, 10) for i in range(15)]

d = {'age': age, 'gender': gender, 'occupation': occupation, 'city': city, 'garden_size': garden_size, 
        'current_planting': current_planting, 'household_income': household_income, 
        'household_size': household_size}
df = pd.DataFrame(data=d)
df
```

```python
age = [random.randint(20, 60) for i in range(15)]
gender = [random.choice([1, 2]) for index in range(15)]
occupation = [random.choice([1, 2, 3, 4, 5]) for index in range(15)]
city = [random.choice([1, 2, 3, 4, 5]) for index in range(15)]
garden_size = [random.randint(0,10) for i in range(15)]
current_planting = [j*random.randint(1, 10) for i, j in zip(range(15), garden_size)]
household_income = [random.randint(2, 10) for i in range(15)]
household_size = [random.randint(1, 10) for i in range(15)]

d = {'age': age, 'gender': gender, 'occupation': occupation, 'city': city, 'garden_size': garden_size, 
        'current_planting': current_planting, 'household_income(1000 scale)': household_income, 
        'household_size': household_size}
df = pd.DataFrame(data=d)
df
```

```python
pd.plotting.parallel_coordinates(df, "age", color=('#000000', '#000000', '#000000', '#556270','#000000', '#000000', 
                                                    '#000000', '#4ECDC4', '#000000', '#000000', '#000000',
                                                    '#C7F464', '#000000', '#000000', '#000000'))
```
