**Generate Input Vector from Tweet**

```python
import nltk
import numpy as np
np.set_printoptions(precision=3)

tweet = "A banana a day keeps the doctor away."
v = {'v0' : 'banana', 'v1' : 'chocolate', 'v2' : 'sweet', 'v3' : 'cookie'}
tweet_words = [word.lower() for word in nltk.word_tokenize(tweet) if word.isalpha()]
x = np.array([tweet_words.count(v[key]) for key in v.keys()])[np.newaxis]
print("Input Vector: ", x)
```

**Layers of the network**

```python
w1 = np.array([[0.5, 0.5, 0, 0], [1, 0.5, 0, 1]])
b1 = np.array([[0], [0]])
w2 = np.array([[0, -1], [1, -1]])
w3 = np.array([[1, 0.5],[1, 0.5], [0.5, 1]])
```

**Sigmoid Function**

```python
def sigmoid(fx):
    return 1 / (1 + np.exp(-fx))

# This function defines a vectorized function which takes a nested sequence of objects or numpy arrays as inputs
# and returns an single or tuple of numpy array as output.
sigmoid_v = np.vectorize(sigmoid)
```

**Display y_predict corresponding to the output labels**

```python
def network_output(y_pred):
    y = ['banana', 'cookie', 'other']
    output = "**Prediction**"
    for index1, index2 in zip(y, y_pred):
        output = output + "\n" + str(index1) + ": " + str(index2)
    return output
```

**Compute forward pass output**

```python
from scipy.special import softmax

z1 = np.dot(w1, x.T) + b1
a1 = sigmoid_v(z1)
z2 = np.dot(w2, a1)
a2 = sigmoid_v(z2)
z3 = np.dot(w3, a2)
y_pred = softmax(z3)
print(network_output(y_pred))
```

**Correct Output Label**

```python
y_true = np.array([[1], [0], [0]])
```

**Log Loss (Cross Entropy)**

```python
from sklearn.metrics import log_loss

cce = log_loss(y_true, y_pred)
print("Cross Entropy/Log Loss: ", np.around(cce, 3))
```

```python
del3 = np.multiply(y_pred-y_true, y_pred*(np.array([[1], [1], [1]])-y_pred))
print(del3)
```

```python
del2 = np.multiply(np.array([[1], [1]])-a2, np.dot(w3.T, del3))
print(del2)
```
