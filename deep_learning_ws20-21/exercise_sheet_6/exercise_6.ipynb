```python
import numpy as np
from scipy.special import softmax

np.set_printoptions(precision=3)

x = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]]) # one-hot encoding of Kevin likes bananas
```

```python
wi = np.array([[0, -1, 0], [-1, 0, 0.5]])
wh = np.array([[0, 1], [1, 0]])
wout = np.array([[-1, 1], [0, -1], [1, 0]])
```

```python
def sigmoid(X):
    return 1/(1+np.exp(-X))
```

```python
a1 = np.add(np.dot(wi, np.array([x[0]]).T), np.dot(wh, sigmoid(np.dot(wi, np.array([x[0]]).T))))
softmax(np.dot(wout, a1))
```

```python
a2 = np.add(np.dot(wi, np.array([x[1]]).T), np.dot(wh, sigmoid(np.dot(wi, np.array([x[1]]).T))))
softmax(np.dot(wout, a2))
```

```python
a3 = np.add(np.dot(wi, np.array([x[2]]).T), np.dot(wh, sigmoid(np.dot(wi, np.array([x[2]]).T))))
softmax(np.dot(wout, a3))
```

```python
a = np.add(np.dot(wi, x), np.dot(wh, sigmoid(np.dot(wi, x))))
softmax(np.dot(wout, a))
```

```python
w_f = np.array([[0, -1, -1, 1, 0], [-1, 1, 0, -1, 1]])
w_i = np.array([[-1, 1, 0, 1, 0], [0, 1, 0, 1, -1]])
w_c = np.array([[-1, 0, 1, 1, 0], [-1, -1, 1, -1, 1]])
w_o = np.array([[-1, 0, 1, 1, 0], [0, 1, 1, -1, 1]])
w_out = np.array([[-1, 1], [0, -1], [0, 0]])
```

```python
h1 = np.array([[0.370], [-0.123]])
c1 = np.array([[0.557], [-0.338]])
```

```python
def column_concatenate(a, b):
    return np.concatenate((a, b), axis=None)[np.newaxis].T
```

```python
f_t = sigmoid(np.dot(w_f, column_concatenate(h1, x[1]))) # bias vector zero
i_t = sigmoid(np.dot(w_i, column_concatenate(h1, x[1]))) # bias vector zero
c_tilde_t = np.tanh(np.dot(w_c, column_concatenate(h1, x[1])))
c_t = np.multiply(f_t, c1) + np.multiply(i_t, c_tilde_t)
o_t = sigmoid(np.dot(w_o, column_concatenate(h1, x[1])))
h_t = np.multiply(o_t, np.tanh(c_t))
```

```python
y_t = np.dot(w_out, o_t)
y_t
```

```python
softmax(y_t)
```
